## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

## Preview Data

Use the following data dictionary to understand the attributes available

|Attribute | Description |
|---------|--------------|
|	mpg	|Miles/(US) gallon|
|	cyl|	Number of cylinders|
|	disp|	Displacement (cu.in.)|
|hp|	Gross horsepower|
|drat|	Rear axle ratio|
|wt|	Weight (1000 lbs)|
|qsec|	1/4 mile time|
|vs|	Engine (0 = V-shaped, 1 = straight)|
|am|	Transmission (0 = automatic, 1 = manual)|
|gear|	Number of forward gears|


# Create basic charts

- Line chart
- Bar chart
- Histogram
- Box plot
- Scatter plot




```python
#Line chart
df.plot(x='carmodel', y='mpg', kind='line', title="MPG by Car")
```

```python
#Bar chart
df.plot(x='carmodel', y='mpg', kind='bar', title="MPG by Car")
```

```python
#Histogram
df['mpg'].plot(kind='hist', bins=10, title="MPG Distribution")
```

```python
#Boxplot
df['mpg'].plot(kind='box', title = 'Miles Per Gallon')
```

```python
#Scatter plot
df.plot.scatter(x='mpg',y='hp',c='wt')
```

```python
df['am'].map({0:'Automatic', 1:'Manual'}).value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue','lightgreen'], title='Distribution of Transmission Types')
```

```python
import matplotlib.pyplot as plt

df.plot(x='carmodel', y='mpg', kind='bar',color=plt.cm.tab20.colors,title="MPG by Car Model")
```

```python
df.plot(x='wt', y='mpg', kind='scatter', title="Weight vs. MPG")
```

```python
df.groupby('cyl')['mpg'].mean().plot(kind='bar', title="Average MPG by Cylinders", rot=0)
```

```python
df['wt'].plot(kind='hist', bins=10,title="Distribution of Car Weights", color='skyblue', edgecolor='black').set_xlabel("Weight")
```
